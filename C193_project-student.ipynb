```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('cardiovascular.csv')
df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of it
grp_age = df.groupby('age')['cholesterol'].mean().reset_index()
grp_age
```

```python
#Task 3 
#Plot a line graph for various age group and their cholesterol 
label = grp_age['age']
value = grp_age['cholesterol']

fig = plt.subplots(figsize=(19,8))

plt.plot(label, value, label = "Average cholesterol level per age" , linewidth=3.0)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend()
plt.title("Average Cholesterol level for per age group", fontsize=20)
plt.show()
```

# Correlation between systolic and diastolic blood pressure

```python
# Diastolic blood pressure - is the pressure in the arteries when the heart rests between beats
# Systolic blood pressure - the pressure in your arteries when your heart beats

#predefine code for image
from IPython.display import Image
Image(filename='blood pressure readings chart.jpg') 
#predefine code end
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it
max_sbp = df.groupby('age')['systolic_blood_pressure'].max().reset_index()
max_sbp
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe project out of it
max_dbp = df.groupby('age')['diastolic_blood_pressure'].max().reset_index()
max_dbp
```

```python
#Task 6
#Plot a line graph to show a Correlation between systolic and diastolic blood pressure
slabel = max_sbp['age']
svalue = max_sbp['systolic_blood_pressure']
fig = plt.subplots(figsize=(19,8))

plt.plot(slabel, svalue, label = "Max Systolic Blood Pressure", linewidth=3.0, color = 'Green')

dlabel = max_dbp['age']
dvalue = max_dbp['diastolic_blood_pressure']

plt.plot(dlabel, dvalue, label="Max Diastolic Blood Pressure", linewidth=3.0, color='red')

plt.xlabel('Age')
plt.xticks(rotation='vertical')

plt.title("Correlation between Systolic and Diastolic Blood Pressure" )
plt.legend()
plt.show()
```
